# Arithmetic mean
Arithmetic mean = mean of the average
or collection of numbers divided by count of numbers 
Always the highest of the means 

statistics.mean
or
.mean()

```python
import statistics
from scipy import stats

statistics.mean(belize_df['price'])
#could also have used this to get the mean
#belize_df['price'].mean()
```

```python
cols=['reviews_per_month', 'calculated_host_listings_count']
belize_df[cols].mean()
```

# Geometric mean
geo mean = a set of numbers whose values are meant to be multiplied together or are exponential in nature
(multiply variables and take the n root) 
Always the middle of the means 

statistics.geometric_mean()
or
scipy.stats.gmean

```python
stats.gmean(belize_df.loc[:,"calculated_host_listings_count"])
```

```python
#Add Geometric Mean column to data frame 
belize_df['Geometric Mean'] = None
#calculate gmean of price and reviews 
belize_df['Geometric Mean'] = stats.gmean(belize_df.iloc[:, 9:11], axis=1)
#print results in dataframe
belize_df.head()
```

# Harmonic mean 
Harmonic mean = the reciprocal of the average of the reciprocals. Always the lowest of the means 

statistics.harmonic_mean() or stats.hmean

```python
#harmonic mean of 
stats.hmean(belize_df.loc[:,"minimum_nights"])
```
